```python
karn=karn[['Date','Lat','Long','Total']]
```

```python

Vis.columns=['Date','Latitude','Longtitude','Total','Hook']
wb.columns=['Date','Latitude','Longtitude','Total']
kerala.columns=['Date','Latitude','Longtitude','Total']
karn.columns=['Date','Latitude','Longtitude','Total']
```

```python
Vis.info()
wb.info()
kerala.info()
karn.info()
```

```python
master=pd.concat([Vis,wb,kerala,karn])
```

```python
master
```

```python
master['Date']=pd.to_datetime(master['Date'],errors='coerce',format='mixed')
```

```python
master.drop_duplicates(keep='first',inplace=True)
```

```python
master=master[~(master.duplicated(subset=['Date','Latitude','Longtitude'],keep=False) & master['Total'].isna())]
```

```python
master.sort_values(['Date','Latitude','Longtitude'])
```

```python
master=master[~((master['Latitude']==0) | (master['Longtitude']==0))]
```

```python
master.dropna(subset=['Latitude','Longtitude'],inplace=True)
```
