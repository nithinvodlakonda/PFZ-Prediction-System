```python
import numpy as np
import pandas as pd 
```

```python
data=pd.ExcelFile(r'C:\Users\vodla\Downloads\Incois\03-Karnataka_Benakappa_Data_2013-14.xlsx')
```

```python
data.sheet_names
```

```python
catch=data.parse('Fish catch')
```

```python
catch.dropna(how='all',inplace=True)
```

```python
catch.drop(columns=['Unnamed: 0','Unnamed: 1','Unnamed: 2','Unnamed: 3','Unnamed: 4'],inplace=True)
```

```python
catch
```

```python
catch.dropna(subset=['Unnamed: 5'],inplace=True)
```

```python
catch
```

```python
catch.columns=catch.iloc[0]
catch=catch[1:]
```

```python
catch
```

```python
catch['Date']=catch['Date'].ffill()
```

```python
cols = list(catch.columns)

cols[2] = 'Lat_start'
cols[3] = 'Long_start'
cols[4] = 'Lat_end'
cols[5] = 'Long_end'

catch.columns = cols
```

```python
catch
```

```python
catch=catch[['Date','Lat_start','Long_start','Lat_end','Long_end','Total']]
```

```python
catch
```

```python
catch['Lat_start'] = catch['Lat_start'].str.replace(r'[oNS]',' ',regex=True)
catch['Long_start'] = catch['Long_start'].str.replace(r'[oEW]',' ',regex=True)
catch['Lat_end'] = catch['Lat_end'].str.replace(r'[oNS]',' ',regex=True)
catch['Long_end'] = catch['Long_end'].str.replace(r'[oEW]',' ',regex=True)
catch['Lat_start'] = catch['Lat_start'].str.replace(r"'"," ",regex=True)
catch['Long_start'] = catch['Long_start'].str.replace(r"'"," ",regex=True)
catch['Lat_end'] = catch['Lat_end'].str.replace(r"'"," ",regex=True)
catch['Long_end'] = catch['Long_end'].str.replace(r"'"," ",regex=True)
```

```python
catch
```

```python
import convert
```

```python
catch['Lat_start']=catch['Lat_start'].apply(convert.convert_coord_3)
catch['Long_start']=catch['Long_start'].apply(convert.convert_coord_3)
catch['Lat_end']=catch['Lat_end'].apply(convert.convert_coord_3)
catch['Long_end']=catch['Long_end'].apply(convert.convert_coord_3)
```

```python
catch
```

```python
catch.info()
```

```python
catch['Lat']= (catch['Lat_start'] + catch['Lat_end'])/2
catch['Long'] = (catch['Long_start'] + catch['Long_end'])/2
```

```python
catch
```

```python
catch.drop(columns=['Lat_start','Long_start','Lat_end','Long_end'],inplace=True)
```

```python
catch
```

```python
catch['Date']=pd.to_datetime(catch['Date'])
```

```python
catch
```

```python
catch.to_csv(r'C:\Users\vodla\Downloads\Incois\Cleaned_karn.csv',index=False)
```
