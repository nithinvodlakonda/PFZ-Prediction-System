```python
final[final['TOTAL_CATCH'].isnull()]
```

```python
np.set_printoptions(threshold=np.inf)
final['TOTAL_CATCH'].unique()
```

```python
final.loc[final['MAJOR_SPECIES'] == 'No Catch','TOTAL_CATCH']= 0
final.loc[final['MAJOR_SPECIES'] == 'Acetus(5), Sciaenids(10), lobster(3no)', 'TOTAL_CATCH'] = 18
final.loc[final['MAJOR_SPECIES'] == 'No fish,','TOTAL_CATCH']=0
```

```python
final['TOTAL_CATCH'] = final['TOTAL_CATCH'].astype(str).str.replace('Nil,','nan' )
final['TOTAL_CATCH'] = final['TOTAL_CATCH'].astype(str).str.replace('Catch(kgs),', '', regex=False)
final['TOTAL_CATCH'] = final['TOTAL_CATCH'].astype(str).str.replace(r'kgs?','', regex=True,case=False)
final['TOTAL_CATCH'] = final['TOTAL_CATCH'].str.replace('No','')
final['TOTAL_CATCH'] = final['TOTAL_CATCH'].str.replace('1ton500','1500')
final['TOTAL_CATCH'] = final['TOTAL_CATCH'].str.replace('-','.')
```

```python
Ton= final[final['TOTAL_CATCH'].str.contains('ton', na=False, case=False)].copy()
Ton['TOTAL_CATCH']=Ton['TOTAL_CATCH'].astype(str).str.replace('ton', '', case=False).astype(float)*1000
```

```python
final.loc[Ton.index, 'TOTAL_CATCH'] = Ton['TOTAL_CATCH']
```

```python
AZ=final[final['TOTAL_CATCH'].str.contains('[A-Za-z]', regex=True, na=False) & (final['TOTAL_CATCH'].astype(str).str.lower() != 'nan')
].copy()
AZ['TOTAL_CATCH']=AZ['TOTAL_CATCH'].astype(str).str.findall(r'\d+\d*').str.join(',')
```

```python
final.loc[AZ.index, 'TOTAL_CATCH'] = AZ['TOTAL_CATCH']
```

```python
spliting=final[pd.to_numeric(final['TOTAL_CATCH'], errors='coerce')>=20000].copy()
```

```python
import split_catch
spliting['TOTAL_CATCH'] = spliting.apply(split_catch.split_catch,axis=1)
```

```python
final.loc[spliting.index, 'TOTAL_CATCH'] = spliting['TOTAL_CATCH']
```

```python
final[final['TOTAL_CATCH'].astype(str).str.contains(',', na=False, regex=True)]
```

```python
Comma=final[final['TOTAL_CATCH'].astype(str).str.contains(',', na=False, regex=True)].copy()
Comma['TOTAL_CATCH'] = Comma['TOTAL_CATCH'].apply( lambda x: round(sum(float(i) for i in x.split(',') if i.strip() != ''), 3))
```

```python
final.loc[Comma.index, 'TOTAL_CATCH'] = Comma['TOTAL_CATCH'].astype(str)
```

```python
final['TOTAL_CATCH'] = pd.to_numeric(final['TOTAL_CATCH'], errors='coerce')
```
